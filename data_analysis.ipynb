Open data. All ECG have a lengh of 186. All the samples are cropped, downsampled and padded with zeroes if necessary to the fixed dimension of 188. Last column has the classification of the ecg : data.iloc[:,-1]

```python

data = prepare_data.load_images_ecg()
mitbih = data['mitbih_train']
ptbdb = data['ptbdb']
```

```python
# Analyzing MITBIH samples - mitbih.shape is (109446, 188) let's preparate the data for the models: Create dataframe to train and ground truth.
mitbid_ecg = mitbih.iloc[:,:186]
mitbid_classification = mitbih.iloc[:,-1]

# See visualy how a ECG looks like
plt.plot(mitbid_ecg.iloc[0].index, mitbid_ecg.iloc[0].values)
```

Check number of different categories in ECG samples (Usually abnormal samples are more rare in this kind of datasets)

```python
np.histogram(mitbid_classification, bins=4)
# 72471 Normal Beats
# 2223 Unknown Beats
# 5788 Ventricular ectopic beats
# 7072 Supraventricular ectopic beats
```
